### Transciptions with Whisper

```python
import os 
import openai

from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv(), override=True)
openai.api_key = os.getenv('OPENAI_API_KEY')

from openai import OpenAI
client = OpenAI()
```

```python
file = 'soad.mp3'
with open(file, 'rb') as audio_file:
    transcript = client.audio.transcriptions.create(
        model='whisper-1',
        file=audio_file
    )
    print(transcript)
```

```python
print(transcript.text)
```

### Translations with Whisper

```python
from openai import OpenAI
client = OpenAI()

with open('german.mp3', 'rb') as audio_file:
    translation = client.audio.translations.create(
        model='whisper-1',
        file=audio_file,
    )

print(translation)
```

```python
from openai import OpenAI
client = OpenAI()

with open('soul.mp3', 'rb') as audio_file:
    translation = client.audio.translations.create(
        model='whisper-1',
        file=audio_file,
    )

print(translation)
```

```python
print(translation.text)
```

### Text to Speech API

```python
# text = 'Viitorul este o carte deschisă, plină de capitole încă nescrise,\
# așteptând strălucirea minților noastre să picteze poveștile lor.'

text = 'it began with the forging of the Great Rings. Three were given to the Elves, immortal, wisest and fairest of all beings. Seven to the Dwarf-Lords, great miners and craftsmen of the mountain halls. And nine, nine rings were gifted to the race of Men, who above all else desire power. For within these rings was bound the strength and the will to govern each race. But they were all of them deceived, for another ring was made. Deep in the land of Mordor, in the Fires of Mount Doom, the Dark Lord Sauron forged a master ring, and into this ring he poured his cruelty, his malice and his will to dominate all life. One ring to rule them all.'
```

```python
cost = len(text)/1000 * 0.15  # cost in cents
cost
```

```python
from openai import OpenAI
client = OpenAI()
response = client.audio.speech.create(
     model="gpt-4o-mini-tts", 
     voice="coral",  
     input=text
)

with open('tts-1.mp3', 'wb') as f:
    f.write(response.content)
```
