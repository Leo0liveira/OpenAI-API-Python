```python
import os 
import openai

from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv(), override=True)
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
from openai import OpenAI
client = OpenAI()
MODEL = 'gpt-4o'

completion = client.chat.completions.create(
    model = MODEL,
    messages = [
        { 'role' : 'system', 'content' : 'Você é um escritor de ficção científica.'},
        { 'role' : 'user', 'content' : 'Escreva uma historia bem curta sobre a vida na Terra no ano 3000.' }
    ]
)

print(completion.choices[0].message.content)
```

## Using local Base64 Images as Input

```python
%pip install ipython
```

```python
from IPython.display import Image, display 
image_path = 'gollum.webp'
display(Image(image_path))
```

```python
import base64
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
```

```python
base64_image = encode_image(image_path)

system_message = 'You are a geek person.'
prompt = 'Who is in this image?'

response = client.chat.completions.create(
    model = MODEL,
    messages = [
        {'role':'system','content':system_message},
        {'role':'user','content':[
            {'type':'text', 'text':prompt},
            {'type':'image_url','image_url': {
                'url':f'data:image/webp;base64,{base64_image}'}
            }
        ]}
    ],
    temperature = 0.0
)
```

```python
print( response.choices[0].message.content)
```

## Using Online Images 

```python
from openai import OpenAI

client = OpenAI()
MODEL = 'gpt-4o'

system_message = 'You respond in English and then translate your response into Japanese'
prompt = 'Write a short and concise description of this image'
image_url = 'https://i.pinimg.com/1200x/53/1b/f2/531bf28ee274611ab3b887c9c301d88a.jpg'

response = client.chat.completions.create(
    model = MODEL,
    messages = [
        {'role':'system','content':system_message},
        {'role':'user','content':[
            {'type':'text', 'text':prompt},
            {'type':'image_url','image_url': {
                'url': image_url}
            }
        ]}
    ],
    temperature = 0.0
)

print( response.choices[0].message.content)
```
